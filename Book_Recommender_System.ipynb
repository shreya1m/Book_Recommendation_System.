# Popularity Based Recommendation System


## This popularity based recommendation system picks the top 50 most popular books with average ratings from books receiving over 150 reviews, offering a curated list of widely adored titles for effortless discovery.

```python
# First merge ratings with books using ISBN column

ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
# Doing group by on ratings_with_name using Book-Title to count ratings of each book

num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Doing group by on ratings_with_name using Book-Title and calculating mean of ratings

avg_rating_df = ratings_with_name.groupby('Book-Title').agg({'Book-Rating': 'mean'}).reset_index()
```

```python
# Renaming Book-Rating as avg_rating

avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
# Here we are merging num_rating_df with avg_ratings on Book-Title

popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df.tail()
```

```python

# Here, we're selecting books reviewed by more than 150 users, sorting them based on their average ratings in descending order,
# and finally displaying the top 50 books,highlighting the most esteemed titles.

popular_df = popular_df[popular_df['num_ratings']>=150].sort_values('avg_rating',ascending=False).head(50)
```

```python
# Merging the popular_df with the books dataframe, as the books contains all the necessary information about each book required for display.

popular_df = popular_df.merge(books,on='Book-Title')
```

```python
# Dropping if there is any duplicates

popular_df = popular_df.drop_duplicates('Book-Title')
```

# Collaborative Filtering Based Recommendation System


## Filtering users who have rated at least 100 books and selecting only those books that have received more than 50 ratings, creating a pivot table with books as index and users as columns. This pivot table facilitates calculating cosine similarity to recommend four most similar books based on user queries.

```python
# Filtering only those users who rated atleast 100 books

x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 100
```

```python
# We need index of those users so we are extracting index of filtered users(x)

padhe_likhe_users = x[x].index
```

```python
padhe_likhe_users
```

```python
# Here we are extracting only users information whose user-id is in padhe_likhe_users

filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
# As we want only famous books so for that we are keeping here only those books who have been rated more than 50 ratings

y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
# To get the names of books we are using .index
famous_books = y[y].index
```

```python
famous_books
```

```python
# Now , here we are extracting only those books that were in famous books from filtered-rating so that now we have final dataframe in which we
# took only those users who rated atleast 100 books and only those books which are rated more than 50 times .

final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# Here we are converting final_rating in pivot table to make it suitable for collaborative filtering and for using cosine similarity
# in index we are taking Book-Title and in columns we will have all User-ID and values that will be filled in this pivot will be Book-Rating

pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
# As so many NA are there so we are filling NA with 0 because NA means no ratings means 0
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
# First book similarity scores against all the books

print(similarity_scores[0])
```

```python
similarity_scores.shape
```

```python
# Main function that calculates details of four similar books based on the Book-Title passed into the function,
# utilizing similarity scores calculated using cosine similarity.


def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
# Testing recommed function which is our main function for giving us 4 recommendations based on our input

recommend('Whispers')
```
